# Basic Plotting with Pandas






### Loading in data for demo

```python
#Import pandas and read in data file

import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/TRIADS_workshops/workshop_data/gdp_europe.csv", index_col=0)
df.head()
```

```python
# Get a quick summary of our data

df.info()
```

```python
# Making a smaller subset of the dataframe so it is easier to see what's going on when we visualize!

df_mini = df.head(5)
df_mini
```

## Generating Basic Graphs from Pandas Dataframes


```python
# Basic bar graph

df_mini.plot(kind='bar')
```

```python
# Basic line chart

df_mini.plot(kind="line", y=df_mini.columns.values)
```

```python
# Basic Box Plot (note--will only work on numeric data)

df_mini.plot(kind='box')
```

```python
# Creating a bar graph with a specific column as the X asix

df_mini.plot(kind='bar', y="1952")
```

```python
df_mini.plot(kind='box', y = "1952")
```

```python
df_mini.plot(kind="line", y = "1952")
```

```python
# Certain charts need more than one parameter to work--like pie charts!

df_mini.plot(kind='pie', y="1952")
```

```python
# Basic Scatterplot (requires x and y axis)

df_mini.plot(kind="scatter", x = "1952", y = "2007")
```

```python
# Basic Box Plot with inverted axes

df_mini.T.plot(kind='box')
```

```python
# Basic line chart (with axes inverted so the columns, instead of the rows, are graphed along the x axis)

df_mini.T.plot(kind="line")
```

```python
df_mini.T.plot(kind="line", y="Belgium")
```

## Exercise

**Make the following graphs using the data in gdp_africa.csv:**

1.   A line graph representing the change in the GDP of Tanzania between 1952 and 2007

2.   A scatterplot representing the relationship between Morocco's GDP in 1972 to it's GDP in 1982

**And now, something harder:**

3.  A boxplot representing the GDP variability of all the countries in Africa starting with "B" (Hint: Use a for loop to filter the countries!)



## Solution

```python
# Line graph representing the change in the GDP of Tanzania between 1952 and 2007

df_africa =  pd.read_csv("/content/drive/MyDrive/TRIADS_workshops/workshop_data/gdp_africa.csv", index_col=0)

df_africa.T.plot(kind="line", y="Tanzania")
```

```python
#  Scatterplot representing the relationship between Morocco's GDP in 1972 to it's GDP in 1982

df_africa.plot(kind="scatter", x="1972", y="1982")
```

```python
# A boxplot representing the GDP variability of all the countries in Africa starting with "B"

b_countries = []

for country in df_africa.index.values:
  if country[0] == "B":
    b_countries.append(country)

df_b_countries = df_africa.loc[b_countries]

df_b_countries.T.plot(kind="box")
```

```python
# Alternative solution to part 3 using an empty dataframe

df_b_countries = pd.DataFrame(columns=df_africa.columns)

for country in df_africa.index.values:
    if country[0] == "B":
        df_b_countries.loc[country] = df_africa.loc[country]

df_b_countries.T.plot(kind="box")
```
